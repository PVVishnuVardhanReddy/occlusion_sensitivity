```python
import tensorflow as tf
from tf_explain.core.occlusion_sensitivity import OcclusionSensitivity
```

```python
model=tf.keras.applications.resnet50.ResNet50(weights="imagenet",include_top=True)
image_path="/content/Screenshot 2024-04-02 at 3.46.27 PM.png"
img = tf.keras.preprocessing.image.load_img(image_path,target_size=(224, 224))
img = tf.keras.preprocessing.image.img_to_array(img)
model.summary()
data = ([img], None)
tabby_cat_class_index = 281
dog = 189
explainer=OcclusionSensitivity()
grid = explainer.explain(data, model, tabby_cat_class_index, 10)
explainer.save(grid, ".", "occlusion_sensitivity_10_cat.png")
grid = explainer.explain(data, model, dog, 20)
explainer.save(grid, ".", "occlusion_sensitivity_10_dog.png")
```
